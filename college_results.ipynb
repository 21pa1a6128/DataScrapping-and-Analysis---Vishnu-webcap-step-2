# Normalising and Preprocessing 

```python
#seperating the cgpas of each sem from the whole data
cgpas=pd.DataFrame([],columns=df.columns)
for i,row in df.iterrows():
    if(row[3]=='0' and row[4]=='0'):
        df.drop([i],inplace=True)
        cgpas.loc[len(cgpas.index)]=row
cgpas['credits'].unique()
```

```python
creditless=df[df.credits=='--']
df=df[df.credits!='--']
```

```python
df[['credits']]=df[['credits']].apply(pd.to_numeric)
```
